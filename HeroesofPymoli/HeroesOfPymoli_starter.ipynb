# Heroes of Pymoli Analysis
* The most profitable and largest target group of the game are males. We see that male players make up 84% of the total number of players with the largest total spend. 
* While the age range of 20-24 has the largest total spend, the average purchase price for the 35-39 age range is signficantly higher by about 50 cents. We can infer that as the age bracket increases, the salary bracket also increases which accounts for higher spend. 
* The most popular games are not necesarily the most profitable. When comparing the popular items dataframe versus the profitable items dataframe, we can see there are a number of cheaper games that are purchased more, but cost less so they are not deemed as profitable.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
#Calculate the total number of players in the dataset
total_players = len(purchase_data_df["SN"].unique())

#Create a dataframe displaying this result
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate the unique items, average price, total purchases, and total revenue
unique_items = len(purchase_data_df["Item ID"].unique())
average_price = purchase_data_df["Price"].mean()
number_of_purchases = purchase_data_df["Purchase ID"].count()
total_revenue = purchase_data_df["Price"].sum()

#Create a summary table with these results
summary_table_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                "Average Price": [average_price],
                                "Number of Purchases": [number_of_purchases],
                                "Total Revenue": [total_revenue]})

#Clean the summary table with correct formatting for Average Price and Total Revenue
summary_table_df["Average Price"] = summary_table_df["Average Price"].map("${:.2f}".format)
summary_table_df["Total Revenue"] = summary_table_df["Total Revenue"].map("${:,.2f}".format)
summary_table_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Sort the list by gender
gender_grouped = purchase_data_df.groupby("Gender")

#Calculate the total count of genders by "SN" to avoid duplicates
gender_count = gender_grouped.nunique()["SN"]

#Calculate the percentage of each gender
gender_percentage = gender_count / total_players

#Create a new dataframe with these results
gender_demographics_df = pd.DataFrame({"Total Count": gender_count,
                                        "Percentage of Players": gender_percentage 
                                       })

#Remove number of index "0,1,2..."
gender_demographics_df.index.name = None

#Format the percentage 
gender_demographics_df["Percentage of Players"] = gender_demographics_df["Percentage of Players"].map("{:,.2%}".format)

#Sort by ascending values
gender_demographics_df = gender_demographics_df.sort_values(["Total Count", "Percentage of Players"], ascending=False)
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calculate the purchase count by gender
purchase_count_by_gender = gender_grouped["Purchase ID"].count()

#Calculate the average purchase price by gender
average_purchase_price_gender = gender_grouped["Price"].mean()

#Calculate the total purchase value by gender
total_purchase_value_gender = gender_grouped["Price"].sum()

#Calculate the average purchase total per person by gender
average_purchase_total_per_person = total_purchase_value_gender / gender_count

#Create a dataframe with these results
purchasing_analysis_gender_df = pd.DataFrame({"Purchase Count": purchase_count_by_gender,
                                             "Average Purchase Price": average_purchase_price_gender,
                                             "Total Purchase Value": total_purchase_value_gender,
                                             "Avg Total Purchase per Person": average_purchase_total_per_person})

#Format the table to account for currency 
purchasing_analysis_gender_df["Average Purchase Price"] = purchasing_analysis_gender_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_gender_df["Avg Total Purchase per Person"] = purchasing_analysis_gender_df["Avg Total Purchase per Person"].map("${:.2f}".format)
purchasing_analysis_gender_df["Total Purchase Value"] = purchasing_analysis_gender_df["Total Purchase Value"].map("${:.2f}".format)

#Set the index to Gender 
purchasing_analysis_gender_df.index.name = "Gender"
purchasing_analysis_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create bins for ages in 4-year intervals and name each bin
bins = [0, 9, 14, 19, 24, 29, 34, 39, 110]
age_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Categorize the existing players using the age bins and add it to the dataframe
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins, labels=age_names, include_lowest=True)

#Create a new variable that sorts the original dataframe by age group
age_grouped_df = purchase_data_df.groupby("Age Group")

#Calculate the total count of players per age group
age_total_count = age_grouped_df["SN"].nunique()

#Calculate the percentage of players per age group
age_percentage = age_total_count / total_players

#Create a summary data frame to hold these results
age_demographics_df = pd.DataFrame({"Total Count": age_total_count,
                                   "Percentage of Players": age_percentage})

#Format the table to round the percentage column to two decimal points
age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:,.2%}".format)
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calculate the purchase count by age group
age_purchase_count = age_grouped_df["Purchase ID"].count()

#Calculate the average purchase price by age group
age_average_purchase = age_grouped_df["Price"].mean()

#Calulcate the total purchase value by age group
age_total_purchase_value = age_grouped_df["Price"].sum()

#Calculate the average total purchase per person by age group
age_avg_purchase_per_person = age_total_purchase_value / age_total_count

#Create a summary dataframe displaying these results
purchasing_analysis_age_df = pd.DataFrame({"Purchase Count": age_purchase_count,
                                          "Average Purchase Price": age_average_purchase,
                                          "Total Purchase Value": age_total_purchase_value,
                                          "Avg Total Purchase per Person": age_avg_purchase_per_person})

#Format the dataframe to account for currency
purchasing_analysis_age_df["Average Purchase Price"] = purchasing_analysis_age_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_age_df["Total Purchase Value"] = purchasing_analysis_age_df["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_age_df["Avg Total Purchase per Person"] = purchasing_analysis_age_df["Avg Total Purchase per Person"].map("${:.2f}".format)
purchasing_analysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Sort the original dataframe by "SN"
screenname_group_df = purchase_data_df.groupby("SN")

#Calculate the purchase count by screen name
sn_purchase_count = screenname_group_df["Purchase ID"].count()

#Calculate the average purchase price by screen name
sn_average_purchase = screenname_group_df["Price"].mean()

#Calculate the total purchase value by screen name
sn_total_purchase = screenname_group_df["Price"].sum()

#Create a summary dataframe holding these results
top_spender_df = pd.DataFrame({"Purchase Count": sn_purchase_count,
                              "Average Purchase Price": sn_average_purchase,
                              "Total Purchase Value": sn_total_purchase})

#Sort the dataframe in descending value to account for top spenders
top_spender_df = top_spender_df.sort_values("Total Purchase Value", ascending=False)

#Format the dataframe to account for currency
top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:.2f}".format)

#Display head of the dataframe, showing only for top 5 spenders
top_spender_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the Item ID, Item Name, and Item Price from original dataframe
items_df = purchase_data_df[["Item ID", "Item Name", "Price"]]

#Group the dataset by both Item ID and Item Name
items_grouped_df = items_df.groupby(["Item ID", "Item Name"])

#Calculate the purchase count for each item ID
item_purchase_count = items_grouped_df["Item ID"].count()

#Calculate the average item price for each item ID
item_average_price = items_grouped_df["Price"].mean()

#Calculate the total purchase value for each item ID
item_total_purchase_value = items_grouped_df["Price"].sum()

#Create a summary dataframe to hold these results
popular_items_df = pd.DataFrame({"Purchase Count": item_purchase_count,
                                   "Item Price": item_average_price,
                                   "Total Purchase Value": item_total_purchase_value})

#Format the table to account for currency 
popular_items_df["Item Price"] = popular_items_df["Item Price"].map("${:.2f}".format)
popular_items_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"].map("${:.2f}".format)

#Sort the summary dataframe in descending order by Purchase Count
popular_items_sorted_df = popular_items_df.sort_values("Purchase Count", ascending=False)

#Display head of the dataframe, showing only for top 5 most popular items
popular_items_sorted_df.head(5)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Repeating steps from earlier section 
#Retrieve the Item ID, Item Name, and Item Price from original dataframe
items_df = purchase_data_df[["Item ID", "Item Name", "Price"]]

#Group the dataset by both Item ID and Item Name
items_grouped_df = items_df.groupby(["Item ID", "Item Name"])

#Calculate the purchase count for each item ID
item_purchase_count = items_grouped_df["Item ID"].count()

#Calculate the average item price for each item ID
item_average_price = items_grouped_df["Price"].mean()

#Calculate the total purchase value for each item ID
item_total_purchase_value = items_grouped_df["Price"].sum()

#Create a summary dataframe to hold these results
popular_items_df = pd.DataFrame({"Purchase Count": item_purchase_count,
                                   "Item Price": item_average_price,
                                   "Total Purchase Value": item_total_purchase_value})

#Sort the dataframe by total purchase value in descending order
profitable_items_df = popular_items_df.sort_values("Total Purchase Value", ascending=False)

#Format the table to account for currency 
profitable_items_df["Item Price"] = profitable_items_df["Item Price"].map("${:.2f}".format)
profitable_items_df["Total Purchase Value"] = profitable_items_df["Total Purchase Value"].map("${:.2f}".format)

#Display head of the dataframe, showing only for top 5 most profitable items
profitable_items_df.head(5)
```
